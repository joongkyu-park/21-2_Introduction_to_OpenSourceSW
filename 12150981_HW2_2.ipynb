```python
# 12150981 박중규 HW#2-2번
# ml-100k 폴더 내에 해당 파일이 있어야합니다.

import pandas as pd
import numpy as np

#데이터 읽어오기
data = pd.read_csv('u.data', delimiter='\t', names = ["user_id", "movi_id", "rating", "timestamp"])
user = pd.read_csv('u.user', delimiter='|', names = ["user_id", "age", "gender", "occupation", "zip_code"])
info = pd.read_csv('u.info')
item = pd.read_csv('u.item',  sep='|', encoding='latin-1', 
                   names = ['movi_id', 'movie_title','release_date','video_release_date','IMDb URL',
                            'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy',
                            'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 
                            'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])
genre = pd.read_csv('u.genre', delimiter='|', names = ["genre", "index"])
occupation = pd.read_csv('u.occupation', names = ["occupation"])

#모든 평가정보가 담긴 'u.data'파일과 사용자의 정보가 담긴'u.user'파일을 outer join시킴('u.data'파일 기준)
data_merged = pd.merge(left = data , right = user, how = "left", on = "user_id")
#모든 평가정보가 담긴 'u.data'파일과 영화 정보가 담긴'u.item'파일을 outer join시킴('u.data'파일 기준)
    #영화제목만 남기고 나머지 열 삭제
movi_names = item.drop(['release_date','video_release_date','IMDb URL',
                            'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy',
                            'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 
                            'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'], axis=1)
data_merged = pd.merge(left = data_merged , right = movi_names, how = "left", on = "movi_id")
```

```python
#2-A-1)for all user

#movi_id, rating, movie_title 열만 남긴 데이터
data_for_all_users = data_merged.drop(["user_id","timestamp", "age", "gender", "occupation", "zip_code"], axis=1).copy()
#movi_id 열을 기준으로 정렬
data_for_all_users.sort_values(by=['movi_id'], axis=0, inplace=True)
#각 영화별로 rating별 개수를 나타내는 counts 열 추가
result_of_data_for_all_users = data_for_all_users.groupby(by=['movi_id', 'movie_title']).size().reset_index(name='counts').copy()
```

```python
#2-B-1)for all user

#각 영화별로 rating별 개수를 나타내는 counts 열 추가 
avg_of_data_for_all_users = data_for_all_users.groupby(by=['movi_id', 'movie_title','rating']).size().reset_index(name='counts').copy()
#각 rating의 개수와 rating을 곱한 값을 나타내는 열 추가
avg_of_data_for_all_users['rating_multiply_counts'] = avg_of_data_for_all_users.apply(lambda x: x.rating * x.counts, axis=1)
#각 영화별로 rating의 총개수를 나타내는 열 추가
avg_of_data_for_all_users['sum_of_counts'] = avg_of_data_for_all_users.groupby(by=['movi_id', 'movie_title']).counts.transform('sum')
#각 영화별로 rating의 총합을 나타내는 열 추가
avg_of_data_for_all_users['sum_of_ratings'] = avg_of_data_for_all_users.groupby(by=['movi_id', 'movie_title']).rating_multiply_counts.transform('sum')
#각 영화별로 rating의 총합을 총개수로 나누어서 rating의 평균을 나타내는 열 추가
avg_of_data_for_all_users['avg_ratings'] = avg_of_data_for_all_users.apply(lambda x: x.sum_of_ratings / x.sum_of_counts, axis=1)
#movi_id, movie_title, avg_ratings 열만 남기기
avg_of_data_for_all_users.drop(['rating','counts','rating_multiply_counts','sum_of_counts','sum_of_ratings'], axis=1, inplace=True)
#중복되는 행 삭제
avg_of_data_for_all_users.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)
```

```python
#ratings의 평점이 높은 순으로 top 100 list를 추출
avg_of_data_for_all_users.sort_values(by=['avg_ratings'], axis=0, ascending=False).head(100)
```
